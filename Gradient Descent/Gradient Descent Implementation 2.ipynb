```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('test_scores.csv')
df
```

```python
def grad_descent(x,y):
    m = b = 0
    learning_rate = 0.00001
    n = len(x)
    plt.scatter(x,y,marker='+',color='red',linewidth='5')
    iteration = 1000
    
    for i in range(iteration):
        y_predicted = m*x + b
        cost = (1/n)*sum([val**2 for val in (y-y_predicted)])
        print('m= {} ,b= {},cost = {},iteration {}'.format(m,b,cost,i))
        plt.plot(x,y_predicted,color='green')
        md = -(1/n)*sum(x*(y-y_predicted))
        bd = -(1/n)*sum((y-y_predicted))
        m -= learning_rate*md
        b -= learning_rate*md
```

```python
x = np.array(df.math)
y = np.array(df.cs)
grad_descent(x,y)
```
