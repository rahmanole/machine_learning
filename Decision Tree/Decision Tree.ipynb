```python
import pandas as pd
import sklearn.preprocessing as prepo
import sklearn.tree
```

```python
df = pd.read_csv('data.csv')
df.head(20)
```

```python
le_company = prepo.LabelEncoder()
le_job = prepo.LabelEncoder()
le_degree = prepo.LabelEncoder()

df['company_n'] = le_company.fit_transform(df['company'])
df['job_n'] = le_job.fit_transform(df['job'])
df['degree_n'] = le_degree.fit_transform(df['degree'])

input_data = df[['company_n','job_n','degree_n']]
output_data = df['salary_more_then_100k']
input_data
```

```python
model = tree.DecisionTreeClassifier()
model.fit(input_data,output_data)
```

```python
model.predict([[0,0,0]])
```
